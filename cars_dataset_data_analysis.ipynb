### Top 10 car make by count

```python
top_10_makes = df['Make'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_makes.plot(kind='bar')
plt.title('Top 10 Car Makes by Count')
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

### Distribution by Origin

```python
top_10_origins = df['Origin'].value_counts()
plt.figure(figsize=(10, 6))
top_10_origins.plot(kind='bar')
plt.title('Origins by Count')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

### Mean engine size based on their Origin

```python
mean_engine_size_by_origin = df.groupby('Origin')['EngineSize'].mean()
plt.figure(figsize=(10, 6))
mean_engine_size_by_origin.plot(kind='bar', color=['blue', 'green'])
plt.title('Mean Engine Size by Origin')
plt.xlabel('Origin')
plt.ylabel('Mean Engine Size')
plt.xticks(rotation=0)
plt.show()
```

### How does the distribution of car types vary between domestic and foreign cars?

```python
origin_car_type_counts = df.groupby(['Origin', 'Type']).size().unstack()
origin_car_type_counts.plot(kind='bar', stacked=True)
plt.title('Distribution of Car Types by Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Cars')
plt.xticks(rotation=0)
plt.legend(title='Car Type')
plt.show()
```

### How does the weight of cars vary with their length?

```python
avg_mpg_by_cylinders = df.groupby('Cylinders')[['MPG_City', 'MPG_Highway']].mean()
avg_mpg_by_cylinders.plot(kind='bar')
plt.title('Average MPG by Cylinder Count')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.xticks(rotation=0)
plt.legend(['MPG City', 'MPG Highway'])
plt.show()
```
